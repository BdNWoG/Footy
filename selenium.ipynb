```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
service = Service(ChromeDriverManager().install())
```

```python
driver = webdriver.Chrome(service=service, options=options)
driver.get('https://www.sportybet.com/int/sport/football')
time.sleep(10)  # Wait for the page to load
content = driver.page_source
soup = BeautifulSoup(content, 'html.parser')
'''
search_box = driver.find_element("name", "q")
search_box.send_keys('ChromeDriver')
search_box.submit()
time.sleep(5) 
'''
# needs to click to check outright odds
# different goal odds and other odds
driver.quit()
```

```python
wrap = soup.find_all('div', class_='match-league-wrap')
for element in wrap:
    print(element)
```

```python
wrap = soup.find_all('span', class_='m-outcome-odds')
odds_values = []
labels = ["Home Win", "Draw", "Away Win", "Over", "Under"]
for element in wrap:
    odds_values.append(float(element.get_text()))
def group_odds(values, group_size=5):
    return [values[i:i + group_size] for i in range(0, len(values), group_size)]

grouped_odds = group_odds(odds_values)
print(grouped_odds)
```
